```python
import csv
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
files = glob.glob("../../out/iterations/*.csv")
print(files)

```

```python
for f in files:
    cost=[]
    fig, ax = plt.subplots(figsize=(6,3))
    with open(f) as csvfile:
        read_csv = csv.reader(csvfile, delimiter=',')
        for row in read_csv:
            cost.append(float(row[0]))
        plt.plot(cost)
        plt.grid()
        plt.title(f)
        plt.ylabel("Cost")
        plt.xlabel("Iterations")
        fname = f.split("\\")[1].replace(".csv",".png")
        plt.savefig("../../out/iterations/"+fname,bbox_inches='tight')
```
