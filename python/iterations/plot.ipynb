```python
import csv
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
files = glob.glob("./*.csv")
print(files)

```

```python
for f in files:
    cost=[]
    fig, ax = plt.subplots(figsize=(9,4))
    with open(f) as csvfile:
        read_csv = csv.reader(csvfile, delimiter=',')
        for row in read_csv:
            cost.append(float(row[0]))
        plt.plot(cost)
        plt.grid()
        plt.title(f)
```
